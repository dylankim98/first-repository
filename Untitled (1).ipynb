```python
# -----------------------------------------
# 1. 라이브러리 로딩
# -----------------------------------------
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm
import warnings
warnings.filterwarnings('ignore')

# -----------------------------------------
# 2. 데이터 경로 & 불러오기
# -----------------------------------------
DATA_PATH = r"C:\Users\kdh98\OneDrive\Desktop\모두의 연구소\5. 시계열"

modify_data = pd.read_csv(
    os.path.join(DATA_PATH, 'sub_upbit_eth_min_tick.csv'),
    index_col=0, parse_dates=True
)

# 간단 시각화
modify_data.loc['2017-11-01':'2017-12-31','close'].plot()
```

```python

# ==========================================================
# ② Using Moving Average 적용하기
# ==========================================================
# ② Using Moving Average 적용하기

momentum_signal = np.sign(
    np.sign(modify_data['close'] - modify_data['close'].rolling(window).mean()) + 1
)

s_momentum_signal = pd.Series(momentum_signal, index=modify_data.index, name='label')

sub_data = modify_data.loc['2017-11-21', 'close']

c_sig = s_momentum_signal.loc['2017-11-21']
c_sig = pd.DataFrame(c_sig)
c_sig.columns = ['label']   # ★ 반드시 해줘야 함 ★

c_sig['color'] = np.where(c_sig['label'] == 1, 'red', 'blue')

plt.figure(figsize=(10,5))
plt.scatter(sub_data.index, sub_data.values, c=c_sig['color'])

```

```python

# ==========================================================
# ③ Local Min-Max 적용하기
# ==========================================================

def get_local_min_max(close, wait=3):
    min_value = close.iloc[0]
    max_value = close.iloc[0]
    n_cnt_min, n_cnt_max = 0, 0
    
    mins, maxes = [], []
    min_idxes, max_idxes = [], []
    b_min_update, b_max_update = False, False
    
    for idx, val in zip(close.index[1:], close.values[1:]):
        if val < min_value:
            min_value = val
            mins.append(min_value)
            min_idxes.append(idx)
            n_cnt_min = 0
            b_min_update = True
        
        if val > max_value:
            max_value = val
            maxes.append(max_value)
            max_idxes.append(idx)
            n_cnt_max = 0
            b_max_update = True
        
        if not b_max_update:
            b_min_update = False
            n_cnt_min += 1
            if n_cnt_min >= wait:
                max_value = min_value
                n_cnt_min = 0
        
        if not b_min_update:
            b_max_update = False
            n_cnt_max += 1
            if n_cnt_max >= wait:
                min_value = max_value
                n_cnt_max = 0
               
    return (
        pd.DataFrame({'min_time': min_idxes, 'local_min': mins}),
        pd.DataFrame({'max_time': max_idxes, 'local_max': maxes})
    )

# 적용
sub_data = modify_data.loc['2017-11-21', 'close']
mins, maxes = get_local_min_max(sub_data, wait=3)

fig, ax = plt.subplots(1, 1, figsize=(10, 5))
ax.plot(sub_data, 'c')
ax.scatter(mins['min_time'], mins['local_min'], c='blue')
ax.scatter(maxes['max_time'], maxes['local_max'], c='red')
ax.set_ylim([sub_data.min()*0.99, sub_data.max()*1.01])
 
```

```python


# ==========================================================
# ④ Trend Scanning 적용하기
# ==========================================================

def t_val_lin_r(close):
    import statsmodels.api as sml
    x = np.ones((close.shape[0], 2))
    x[:, 1] = np.arange(close.shape[0])
    ols = sml.OLS(close, x).fit()
    return ols.tvalues[1]

look_forward_window = 60
min_sample_length = 5

molecule = modify_data['2017-11-01':'2017-11-30'].index
label = pd.DataFrame(index=molecule, columns=['t1', 't_val', 'bin'])
tmp_out = []

for ind in tqdm(molecule):
    subset = modify_data.loc[ind:, 'close'].iloc[:look_forward_window]
    if look_forward_window > subset.shape[0]:
        continue
    
    tmp_subset = pd.Series(index=subset.index[min_sample_length-1:subset.shape[0]-1])
    tval = []

    for forward_window in np.arange(min_sample_length, subset.shape[0]):
        df = subset.iloc[:forward_window]
        tval.append(t_val_lin_r(df.values))

    tmp_subset.loc[tmp_subset.index] = np.array(tval)

    idx_max = tmp_subset.replace([-np.inf, np.inf, np.nan], 0).abs().idxmax()
    tmp_t_val = tmp_subset[idx_max]

    tmp_out.append([tmp_subset.index[-1], tmp_t_val, np.sign(tmp_t_val)])

label.loc[molecule] = np.array(tmp_out)

label['t1'] = pd.to_datetime(label['t1'])
label['bin'] = pd.to_numeric(label['bin'], downcast='signed')

# 시각화
sub_data = modify_data.loc['2017-11-21', 'close']
c_sig = label['bin'].loc['2017-11-21']
c_sig = pd.DataFrame(c_sig)
c_sig['color'] = np.where(c_sig['bin']==1, 'red', 'blue')

fig, ax = plt.subplots(1, 1, figsize=(10, 5))
ax.scatter(sub_data.index, sub_data.values, c=c_sig['color'])
```

```python
# ==========================================================
# 2. Feature Engineering (2만 행만 사용 / 네 로컬 전용)
# ==========================================================

import sys, os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import ta
import warnings
warnings.filterwarnings("ignore")

# ----------------------------------------------------------
# ① 중요 모듈 경로 설정
# ----------------------------------------------------------
sys.path.append(r"C:\Users\kdh98\OneDrive\Desktop\모두의 연구소\5. 시계열")
import importance as imp   # importance.py


# ----------------------------------------------------------
# ② 파일 로딩
# ----------------------------------------------------------
DATA_PATH = r"C:\Users\kdh98\OneDrive\Desktop\모두의 연구소\5. 시계열"

# 가격 데이터
csv_path = os.path.join(DATA_PATH, "sub_upbit_eth_min_tick.csv")
df_price = pd.read_csv(csv_path, index_col=0, parse_dates=True)

# Feature + Label 포함된 PKL
pkl_path = os.path.join(DATA_PATH, "sub_upbit_eth_min_feature_labels.pkl")
df_label_feature = pd.read_pickle(pkl_path)

print("CSV:", df_price.shape)
print("PKL:", df_label_feature.shape)

# label 생성
df_label_feature["bin"] = np.sign(df_label_feature["t_value"])


# ==========================================================
# 2만 행만 사용하도록 자르기 (중요)
# ==========================================================
MAX_ROWS = 20000
df_label_feature = df_label_feature.iloc[:MAX_ROWS]
df_price = df_price.loc[df_label_feature.index]

print("===== After Trimming to 20,000 rows =====")
print("CSV:", df_price.shape)
print("PKL:", df_label_feature.shape)


# ----------------------------------------------------------
# ③ Label 컬럼 정의
# ----------------------------------------------------------
LABEL_COL = "bin"
y = df_label_feature[[LABEL_COL]]


# ----------------------------------------------------------
# ④ Feature 정의 (PKL 내부 Feature)
# ----------------------------------------------------------
OHLCV = ["open", "high", "low", "close", "volume"]

# PKL에서 OHLCV, label, t1, t_value 제거 → 나머지가 feature
feature_cols = [
    c for c in df_label_feature.columns 
    if c not in OHLCV + [LABEL_COL, "t1", "t_value"]
]

X_feature = df_label_feature[feature_cols]

print("기존 PKL Feature 개수:", len(feature_cols))


# ----------------------------------------------------------
# ⑤ CSV에서 OHLCV만 가져오기
# ----------------------------------------------------------
X_price = df_price.loc[X_feature.index, OHLCV]


# ----------------------------------------------------------
# ⑥ 최종 Feature 합치기
# ----------------------------------------------------------
X = pd.concat([X_price, X_feature], axis=1)

print("최종 Feature Shape:", X.shape)
print("최종 Label Shape:", y.shape)


# ==========================================================
# ⑦ Technical Indicators 추가
# ==========================================================
df_ = X.copy()

open, high, low, close, volume = OHLCV

# -------------------------------
# Volume Indicators
# -------------------------------
df_["volume_cmf"] = ta.volume.ChaikinMoneyFlowIndicator(
    high=df_[high], low=df_[low], close=df_[close], 
    volume=df_[volume], window=20, fillna=False
).chaikin_money_flow()

df_["volume_fi"] = ta.volume.ForceIndexIndicator(
    close=df_[close], volume=df_[volume], window=15,
    fillna=False
).force_index()

df_["volume_mfi"] = ta.volume.MFIIndicator(
    high=df_[high], low=df_[low], close=df_[close], 
    volume=df_[volume], window=15, fillna=False
).money_flow_index()

df_["volume_sma_em"] = ta.volume.EaseOfMovementIndicator(
    high=df_[high], low=df_[low], volume=df_[volume], 
    window=15, fillna=False
).sma_ease_of_movement()

df_["volume_vpt"] = ta.volume.VolumePriceTrendIndicator(
    close=df_[close], volume=df_[volume], fillna=False
).volume_price_trend()


# -------------------------------
# Volatility Indicators
# -------------------------------
df_["volatility_atr"] = ta.volatility.AverageTrueRange(
    high=df_[high], low=df_[low], close=df_[close],
    window=10, fillna=False
).average_true_range()

df_["volatility_ui"] = ta.volatility.UlcerIndex(
    close=df_[close], window=15, fillna=False
).ulcer_index()


# -------------------------------
# Trend Indicators
# -------------------------------
df_["trend_macd_diff"] = ta.trend.MACD(
    close=df_[close], window_slow=25,
    window_fast=10, window_sign=9, fillna=False
).macd_diff()

df_["trend_adx"] = ta.trend.ADXIndicator(
    high=df_[high], low=df_[low], close=df_[close],
    window=15, fillna=False
).adx()

df_["trend_trix"] = ta.trend.TRIXIndicator(
    close=df_[close], window=15, fillna=False
).trix()

df_["trend_mass_index"] = ta.trend.MassIndex(
    high=df_[high], low=df_[low], window_fast=10,
    window_slow=25, fillna=False
).mass_index()

df_["trend_dpo"] = ta.trend.DPOIndicator(
    close=df_[close], window=20, fillna=False
).dpo()

# AroonIndicator (네 ta 버전에 맞는 형태)
df_["trend_aroon_ind"] = ta.trend.AroonIndicator(
    high=df_[high], low=df_[low], window=20, fillna=False
).aroon_indicator()


# -------------------------------
# Momentum Indicators
# -------------------------------
df_["momentum_rsi"] = ta.momentum.RSIIndicator(
    close=df_[close], window=15, fillna=False
).rsi()

df_["momentum_wr"] = ta.momentum.WilliamsRIndicator(
    high=df_[high], low=df_[low], close=df_[close],
    lbp=15, fillna=False
).williams_r()


# -------------------------------
# Return / Volatility
# -------------------------------
windows_mom = [5, 10, 20]
windows_std = [30]

for i in windows_mom:
    df_[f"ret_{i}"] = df_[close].pct_change(i).round(6)
    df_[f"vol_change_{i}"] = df_[volume].pct_change(i).round(6)

for i in windows_std:
    df_[f"std_{i}"] = df_[close].rolling(i).std()
    df_[f"vol_std_{i}"] = df_[volume].rolling(i).std()


# ----------------------------------------------------------
# 최종 Dataset 정리
# ----------------------------------------------------------
df_final = df_.join(y).dropna()

X_final = df_final.drop(columns=[LABEL_COL])
y_final = df_final[LABEL_COL]
```

# 3. modeling

```python
# ==========================================================
# 3. 프로젝트 : Model Training (최종 완성본)
# ==========================================================

import datetime
import sys
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score,
    confusion_matrix, roc_auc_score, roc_curve
)
from sklearn.ensemble import RandomForestClassifier, BaggingClassifier

# ======  로컬 PKFold 파일 경로 등록 ======
sys.path.append(r"C:\Users\kdh98\OneDrive\Desktop\모두의 연구소\5. 시계열")
from pkfold import PKFold   

# ====== 파일 경로 ======
DATA_PATH = r"C:\Users\kdh98\OneDrive\Desktop\모두의 연구소\5. 시계열"
pkl_file = os.path.join(DATA_PATH, "sub_upbit_eth_min_feature_labels.pkl")

# ====== 데이터 불러오기 ======
df_data = pd.read_pickle(pkl_file)

# Trend Scanning 라벨(bin) 생성
df_data["bin"] = np.sign(df_data["t_value"])

print(df_data["bin"].value_counts())

# ----------------------------------------------------------
# Train/Test Split
# ----------------------------------------------------------

train_ratio, test_ratio = 0.7, 0.2
n_train = int(len(df_data) * train_ratio)
n_test  = int(len(df_data) * test_ratio)

# X, y 분리
X = df_data.iloc[:, 5:-1]
y = df_data["bin"]

# Scaling
sc = StandardScaler()
X_sc = sc.fit_transform(X)

train_x = X_sc[:n_train]
test_x  = X_sc[-n_test:]
train_y = y.iloc[:n_train]
test_y  = y.iloc[-n_test:]

# DataFrame 형태 복구
train_x = pd.DataFrame(train_x, index=train_y.index, columns=X.columns)
test_x  = pd.DataFrame(test_x, index=test_y.index, columns=X.columns)

# 학습 데이터 1000개만 사용
train_x = train_x[:1000]
train_y = train_y[:1000]

# ----------------------------------------------------------
# ② Purged K-Fold 적용
# ----------------------------------------------------------

n_cv = 4
t1 = pd.Series(train_y.index.values, index=train_y.index)
cv = PKFold(n_cv, t1, 0)

# ----------------------------------------------------------
# ③ Model 적용하기
# ----------------------------------------------------------

# RandomForest
rfc = RandomForestClassifier(class_weight="balanced")

# sklearn 최신 버전: estimator 사용
bag_rfc = BaggingClassifier(estimator=rfc)

# 최신 문법 파라미터
bc_params = {
    "n_estimators": [5],
    "max_features": [0.5],
    "estimator__max_depth": [3],
    "estimator__max_features": [None, "auto"],
    "estimator__min_samples_leaf": [3],
    "bootstrap_features": [False, True]
}

gs_rfc = GridSearchCV(
    bag_rfc,
    bc_params,
    cv=cv,
    n_jobs=-1,
    verbose=1
)

# ====== Training ======
gs_rfc.fit(train_x, train_y)

# ====== Best model ======
gs_rfc_best = gs_rfc.best_estimator_
gs_rfc_best.fit(train_x, train_y)

# ====== Prediction ======
pred_y = gs_rfc_best.predict(test_x)
prob_y = gs_rfc_best.predict_proba(test_x)[:, 1]

# ----------------------------------------------------------
# 성능 평가 (multiclass 대응)
# ----------------------------------------------------------

confusion = confusion_matrix(test_y, pred_y)
accuracy  = accuracy_score(test_y, pred_y)
precision = precision_score(test_y, pred_y, average="weighted")
recall    = recall_score(test_y, pred_y, average="weighted")

print('================= confusion matrix ====================')
print(confusion)
print('=======================================================')
print(f'정확도: {accuracy:.4f}')
print(f'정밀도(weighted): {precision:.4f}')
print(f'재현율(weighted): {recall:.4f}')

# ----------------------------------------------------------
# ROC Curve (1 vs rest)
# ----------------------------------------------------------

try:
    fpr, tpr, thresholds = roc_curve(test_y, prob_y, pos_label=1.0)
    auc = roc_auc_score(test_y, prob_y)

    plt.plot(fpr, tpr, linewidth=2)
    plt.plot([0, 1], [0, 1], 'k--')
    plt.xlabel("fpr")
    plt.ylabel("tpr")
    plt.title(f"ROC Curve (AUC = {auc:.4f})")
    plt.show()

    print(f"AUC: {auc:.4f}")

except:
    print("\nROC Curve는 binary(또는 one-vs-rest)에서만 계산됩니다.")
```
